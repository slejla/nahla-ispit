Za dostavljeni Iris dataset odaberite klasifikacijski model kojim ce biti moguce sto bolje na osnovu features u datasetu predvidjeti klasu biljke. 


Zadatak ce biti ocijenjen po sljedecim kriterijima:

*   Kod (urednost, logicnost, modularnost)

*   Odabir features, testiranje, metrike (dodatni bodovi za vizualizacije)


```python
import pandas as pd

## ispisujemo tabelu, mijenjamo prazna mjesta s nulom i provjeravamo da li je tabela spremna 
df1=pd.read_csv(r"C:\Users\Home\Desktop\Iris.csv",index_col=0)
df1.head()
df= df1.fillna(0)
df.info()


## određujemo feature i label
x=df.iloc[:, :-1].values
y=df.iloc[:, 4].values

## dijelimo dataset u manje dataset-ove kako bismo lakše radili analizu
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x,y, test_size=0.20)

## mogućnost normalizacije - ako uradimo normalizaciju, na confusion matrici ćemo primijetiti da nestaju vrijednosti koje se odnose na rezultate koji su rezidual
#from sklearn.preprocessing import StandardScaler
#scaler=StandardScaler()
#scaler.fit(x_train)

#x_train=scaler.transform(x_train)
#x_test=scaler.transform(x_test)

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train, y_train)


y_pred = classifier.predict(x_test)

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

## dijagonalni brojevi su broj predviđenih događaja koji su jednaki istinitim.
##Brojevi koji nisu u dijagonal su oni koje je klasifikator pogrešno označio. 
##Vidimo da je to u našem slučaju mala vrijednost što pokazuje dobro predviđanje.
## u slučaju kada se izvrši normalizacija, rezultati su "savršeni" u smislu tačnosti predviđanja.

## vizalizacija podataka
import seaborn as sns
sns.set_palette('husl')
g = sns.violinplot(y='Species', x='SepalLengthCm', data=df, inner='quartile')
plt.show()
g = sns.violinplot(y='Species', x='SepalWidthCm', data=df, inner='quartile')
plt.show()
g = sns.violinplot(y='Species', x='PetalLengthCm', data=df, inner='quartile')
plt.show()
g = sns.violinplot(y='Species', x='PetalWidthCm', data=df, inner='quartile')
plt.show()

## prema grafovima ispod vidimo distribuciju datih osobina za svaku od vrsta irisa. 
##Također, prema grafovima možemo primijetiti da osobine vrste iris-setosa odstupaju od osobina druge dvije.
```
